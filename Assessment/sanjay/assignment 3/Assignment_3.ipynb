# ● Download the Dataset

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
import os
```

```python
batch_size = 16
```

# ● Image Augmentation

```python
data_aug = Sequential(
  [
    layers.RandomFlip("horizontal",input_shape=(180, 180, 3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)
```

```python
os.listdir("/content/flowers")
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
  "/content/flowers",
  validation_split=0.25,
  subset="training",
  seed=120,
  image_size=(180, 180),
  batch_size=batch_size)
```

```python
val_data_set = tf.keras.utils.image_dataset_from_directory(
 "/content/flowers",
  validation_split=0.25,
  subset="validation",
  seed=120,
  image_size=(180, 180),
  batch_size=batch_size)
```

```python
class_names = train_data.class_names
```

```python
plt.figure(figsize=(15, 15))
for images, labels in train_data.take(1):
  for i in range(2):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
```

```python
normalization_layer = layers.Rescaling(1./255)
```

```python
dataset_normalized = train_data.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(dataset_normalized))
first_image = image_batch[0]
print(np.min(first_image), np.max(first_image))
```

# ● Create Model

# ● Add Layers (Convolution,MaxPooling,Flatten,Dense-(Hidden Layers),Output)

```python
num_classes = len(class_names)

model = Sequential([
  data_aug,
  layers.Rescaling(1./255, input_shape=(180, 180, 3)),
  layers.Conv2D(16, 3, activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3,activation='relu'),
  layers.Conv2D(32, 3,activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])
```

# ● Compile The Model

```python
# compiling model with categorical cross entropy and adam optimizer
model.compile(optimizer='adam',
loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
metrics=['accuracy'])
```

# ● Fit The Model

```python
epochs=10
history = model.fit(train_data,validation_data=val_data_set,epochs=epochs)
```

# ● Save The Model

```python
model.save("./flowers.h5")
```

```python
model.load_weights('./flowers.h5')
```
